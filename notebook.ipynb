```python
from retell import Retell

def getCallObject(call_id):    
    client = Retell(
        api_key=os.getenv("RETELL_API_KEY"),
    )
    call_response = client.call.retrieve(
        call_id
    )
    return call_response

```

```python
call_obj = getCallObject("call_c79b56fd1b7e28f691b81909545")
pprint.pprint(call_obj.call_analysis.custom_analysis_data)
```
